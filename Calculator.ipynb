```python
import tkinter as tk 
from tkinter import*

cal=Tk()
cal.title("Calculator")
operator=""
text_input=StringVar()

def btnClick(numbers):
    global operator
    operator = operator + str(numbers)
    text_input.set(operator)
    
def btnClearDisplay():
    global operator
    operator = ""
    text_input.set("")
    
def btnEqualsInput():
    global operator
    ans = str(eval(operator))
    text_input.set(ans)
    
def fail():
    numbers/0
    try:
        fail()
    except:
        print("Exception ZeroDivisionError arise")
    

    

numEntry=Entry(cal,bd = 30,font=("arial",20,"bold"), insertwidth=4, bg="light grey",justify="right", textvariable=text_input)
numEntry.grid(columnspan=4)

but1=Button(cal, padx=16 ,bd=8, text="7",command=lambda:btnClick(7),font=("arial",20,"bold"), fg="black",bg ="Pink")
but1.grid(row=1,column=0)

but2=Button(cal,padx=16 ,bd=8,  text="8",font=("arial",20,"bold"),command=lambda:btnClick(8), fg="black",bg ="Pink")
but2.grid(row=1,column=1)

but3=Button(cal, padx=16 , bd=8, text="9",font=("arial",20,"bold"),command=lambda:btnClick(9), fg="black",bg ="Pink")
but3.grid(row=1,column=2)

but4=Button(cal,padx=16 , bd=8, text="+",font=("arial",20,"bold"),command=lambda:btnClick("+"), fg="black",bg ="Pink")
but4.grid(row=1,column=3)

but5=Button(cal, padx=16 ,bd=8,  text="4",font=("arial",20,"bold"),command=lambda:btnClick(4), fg="black",bg ="Pink")
but5.grid(row=2,column=0)

but6=Button(cal,padx=16 , bd=8, text="5",font=("arial",20,"bold"),command=lambda:btnClick(5), fg="black",bg ="Pink")
but6.grid(row=2,column=1)

but7=Button(cal, padx=16 ,bd=8,  text="6",font=("arial",20,"bold"),command=lambda:btnClick(6), fg="black",bg ="Pink")
but7.grid(row=2,column=2)

but8=Button(cal,padx=16 ,bd=8,  text="-",font=("arial",20,"bold"),command=lambda:btnClick("-"), fg="black",bg ="Pink")
but8.grid(row=2,column=3)

but9=Button(cal,padx=16 , bd=8,  text="1",font=("arial",20,"bold"),command=lambda:btnClick(1), fg="black",bg ="Pink")
but9.grid(row=3,column=0)

but10=Button(cal,padx=16 ,bd=8,  text="2",font=("arial",20,"bold"),command=lambda:btnClick(2), fg="black",bg ="Pink")
but10.grid(row=3,column=1)

but11=Button(cal,padx=16 ,bd=8,  text="3",font=("arial",20,"bold"),command=lambda:btnClick(3), fg="black",bg ="Pink")
but11.grid(row=3,column=2)

but12=Button(cal,padx=16 ,bd=8,  text="*",font=("arial",20,"bold"),command=lambda:btnClick("*"), fg="black",bg ="Pink")
but12.grid(row=3,column=3)

but13=Button(cal,padx=16 ,bd=8,  text="C",font=("arial",20,"bold"),command=btnClearDisplay, fg="black",bg ="Pink")
but13.grid(row=4,column=0)

but14=Button(cal,padx=16 , bd=8, text="0",font=("arial",20,"bold"),command=lambda:btnClick(0), fg="black",bg ="Pink")
but14.grid(row=4,column=1)

but15=Button(cal,padx=16 ,bd=8,  text="=",font=("arial",20,"bold"),command=btnEqualsInput, fg="black",bg ="Pink")
but15.grid(row=4,column=2)

but16=Button(cal,padx=16 , bd=8, text="/",font=("arial",20,"bold"),command=lambda:btnClick("/"), fg="black",bg ="Pink")
but16.grid(row=4,column=3)




cal.mainloop()
```
